```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("HousePrice_dataset.csv")
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
df['area_sqft'] = df['area_in_marla'] * 272.25
```

```python
df = df[['area_sqft', 'bedrooms', 'baths', 'price']].dropna()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['area_sqft'], y=df['price'])
plt.title("Price vs. Area (sq ft)")
plt.xlabel("Area (sq ft)")
plt.ylabel("Price")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['bedrooms'], y=df['price'])
plt.title("Price Comparison by Number of Bedrooms")
plt.xlabel("Bedrooms")
plt.ylabel("Price")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['baths'], y=df['price'])
plt.title("Price Comparison by Number of Bathrooms")
plt.xlabel("Bathrooms")
plt.ylabel("Price")
plt.grid(True)
plt.tight_layout()
plt.show()
```
